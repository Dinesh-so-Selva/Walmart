```python
import pandas as pd

file_path = r'C:\Users\user\Desktop\walmart\Walmart_customer_purchases.csv'
df = pd.read_csv(file_path)
```

```python
print(df.info())
print(df.describe())
print(df['City'].value_counts())
print(df['Repeat_Customer'].value_counts())
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df['Month'] = df['Purchase_Date'].dt.to_period('M')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
monthly_sales = df.groupby('Month')['Bill Amount'].sum()
monthly_sales.plot(kind='line', title='Monthly Revenue')
```

```python
sns.barplot(x='Category', y='Bill Amount', data=df)
```

```python
sns.boxplot(x='Repeat_Customer', y='Bill Amount', data=df)
```
